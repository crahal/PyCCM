```python
import os
import seaborn as sns
from tqdm import tqdm
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from data_loaders import get_fertility, get_lifetables_ex
from figures_static import plot_national_fert_static
import numpy as np

mpl.rcParams['font.family'] = 'Helvetica'
```

```python


df_mid = pd.read_csv('/home/jinx/Dropbox/PyCCM/results/projections/age_structures_df_T/mid_omissions/age_structures_df_TEEVV.csv',
                     index_col=0)
df_low = pd.read_csv('/home/jinx/Dropbox/PyCCM/results/projections/age_structures_df_T/low_omissions/age_structures_df_TEEVV.csv',
                     index_col=0)
df_high = pd.read_csv('/home/jinx/Dropbox/PyCCM/results/projections/age_structures_df_T/high_omissions/age_structures_df_TEEVV.csv',
                     index_col=0)


import matplotlib.pyplot as plt

totals_mid = (
    df_mid.groupby(["year", "DPTO_NOMBRE"])["VALOR_corrected"]
      .sum()
      .reset_index()
)
totals_low = (
    df_low.groupby(["year", "DPTO_NOMBRE"])["VALOR_corrected"]
      .sum()
      .reset_index()
)

totals_high = (
    df_high.groupby(["year", "DPTO_NOMBRE"])["VALOR_corrected"]
      .sum()
      .reset_index()
)

mask_low = (totals_low["DPTO_NOMBRE"] == "total_nacional")
mask_high = (totals_high["DPTO_NOMBRE"] == "total_nacional")
mask_mid = (totals_mid["DPTO_NOMBRE"] == "total_nacional")



plt.plot(totals_low.loc[mask_low, "year"], totals_low.loc[mask_low, "VALOR_corrected"], marker="o")
plt.plot(totals_high.loc[mask_low, "year"], totals_high.loc[mask_high, "VALOR_corrected"], marker="o")
plt.plot(totals_mid.loc[mask_low, "year"], totals_mid.loc[mask_mid, "VALOR_corrected"], marker="o")
```
